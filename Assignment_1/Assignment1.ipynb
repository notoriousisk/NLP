## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from collections import Counter, defaultdict

class BPETokenizer:
    def __init__(self, vocab_size: int) -> None:
        """Initialize the BPETokenizer object with the specified vocabulary size."""
        self.vocab_size = vocab_size
        self.mergings: list[tuple[str, str]] = []

    def train(self, corpus: List[str]) -> None:
        """Train the tokenizer of  Byte Pair Encoding algorithm using the provided corpus"""

        corpus_tokens: list[list[str]] = list(map(lambda text: list(text) + ['</w>'], corpus))
        vocab: set[str] = set(' '.join(corpus))

        while len(vocab) < self.vocab_size:
            pair_counter = Counter()
            for tokens in corpus_tokens:
                pairs = [(tokens[i], tokens[i + 1]) for i in range(len(tokens) - 1)]
                pair_counter.update(pairs)

            if not pair_counter:
                break

            # The new merge is the most common pair in the corpus
            (new_merge, _count), = pair_counter.most_common(n=1)
            self.mergings.append(new_merge)
            new_token = ' '.join(new_merge)
            vocab.add(new_token)

            # Apply the new merge to the corpus
            corpus_tokens = [self._merge_pairs(new_merge, tokens) for tokens in corpus_tokens]

    def tokenize(self, text: str) -> List[str]:
        """Tokenize the given text using the trained vocabulary"""
        tokens = list(text)
        for merge in self.mergings:
            tokens = self._merge_pairs(merge, tokens)
        return tokens

    def _merge_pairs(self, merge: tuple[str, str], tokens: list[str]) -> list[str]:
        """Merge tokens based on the given pair and rule."""
        vocabulary = []
        merged = False
        for current, next_token in zip(tokens, tokens[1:] + [None]):
          if merged:
              merged = False
              continue
          if (current, next_token) == merge:
              vocabulary.append(current + next_token)
              merged = True
          else:
              vocabulary.append(current)
        return vocabulary
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
import zipfile
archive = 'Assignment1_data.zip'
unzip_folder = './Assignment1_data/'
with zipfile.ZipFile(archive, 'r') as zip_ref:
    zip_ref.extractall(unzip_folder)
```

```python
import glob

# Load your tweets dataset into a list
files = glob.glob('./Assignment1_data/file*')
tweets_data = []
for f in files:
    with open(f, 'r') as file:
        tweets_data.extend(file.readlines())

print('Tweets number:', len(tweets_data))
```

```python
# Define a range of vocab_size values
vocab_sizes = [100, 250, 500, 750, 1000]

# Track average token length for each vocab_size
avg_lengths = []

for vocab_size in vocab_sizes:
    print(f'vocabulary size {vocab_size}')
    tokenizer = BPETokenizer(vocab_size=vocab_size)
    tokenizer.train(tweets_data)

    # Tokenize tweets and calculate average length
    tokens = tokenizer.tokenize(' '.join(tweets_data))
    token_lengths = [len(tokenizer.tokenize(tweet)) for tweet in tweets_data]
    avg_lengths.append(sum(token_lengths) / len(token_lengths))
    print("tokens:", tokens[:100])
```

```python
import matplotlib.pyplot as plt

# Plot the results
plt.plot(vocab_sizes, avg_lengths, marker='o')
plt.xlabel('Vocabulary Size')
plt.ylabel('Average Token Length')
plt.title('Dependency of Average Token Length on Vocabulary Size')
plt.show()
```


The graph illustrates that a larger vocabulary, in this case 1000, leads to shorter tokenized tweets. This preference stems from balancing space reduction while maintaining meaning, as a greater vocabulary size allows the tokenizer to capture diverse and specific words, enhancing accuracy and comprehensive tokenization of tweets.
